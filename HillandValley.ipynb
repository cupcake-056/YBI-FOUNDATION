HILL AND VALLEY PREDICTION USING LOGISTICS REGRESSION.

OBJECTIVE

Develop a machine learning model to accurately classify geographical locations as either 'hill' or 'valley' using logistic regression.

Data Source

The dataset can be obtained from the UCI Machine Learning Repository or similar sources

IMPORT LIBRARY

```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

IMPORT DATA

```python
df=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Hill%20Valley%20Dataset.csv")
```

DESCRIBE DATA

```python
#head
df.head()
```

```python
#information about dataset
df.info()
```

```python
#display summary statistics
df.describe()
```

```python
#shape
df.shape
```

```python
#column
df.columns
```


Define Target Variable (y) and Feature Variables (X)

```python
#Target variable(y)
y = df['Class']
y.shape
```

```python
#Feature Variable(x)
x = df.drop(['Class'], axis=1)
x.shape
```


Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

Modeling

```python
model = LogisticRegression(max_iter=5500)
model.fit(x_train,y_train)
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
score
```

Prediction

```python
x_new = df.sample(1)
x_new
```

```python
x_new = x_new.drop(['Class'], axis=1)
x_new
```

```python
prediction = model.predict(x_new)
prediction
```
